```python
import networkx as nx
import ringity
```

```python
diG = nx.read_edgelist("finding rings/faa routes/edges.csv", delimiter=",", comments="#", nodetype=int, create_using=nx.DiGraph)
G = diG.to_undirected()
nx.info(G)
```

```python
dgm = ringity.diagram(G, metric="net_flow")
ringity.plot_dgm(dgm)
dgm.ring_score()
```
