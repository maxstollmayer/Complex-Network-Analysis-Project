# Example Notebook

```python
import networkx as nx
import ringity as rng
from matplotlib import pyplot as plt
from layout import Layout
```

## Create and Plot Network Model

```python
N = 2**10   # number of nodes
a = 0.25    # activity duration
beta = 0.8  # randomness
rho = 0.1   # expected density

G = rng.network_model(N=N, a=a, beta=beta, rho=rho)
rng.plot_nx(G)
```

```python
# create and plot its persistance diagram
dgm = rng.diagram(G)
rng.plot_dgm(dgm)
```

```python
# plot persistence sequence (hole sizes)
seq = dgm.sequence
plt.plot(range(len(seq)), seq)
plt.show()
```

```python
# print ring score
dgm.ring_score()
```

## Create Layout and 3D Preview

```python
# create 3D positions
pos = nx.spring_layout(G, dim=3)
```

```python
# create layout
layout = Layout(G, pos)
```

```python
# print node table
layout.node_table
```

```python
# print edge table
layout.edge_table
```
