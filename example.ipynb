```python
import networkx as nx
from layout import Layout
```

```python
# example graph and positions
G = nx.erdos_renyi_graph(10, 0.25)
pos = nx.spring_layout(G, dim=3)

# new layout instance
layout = Layout(G, pos)
```

```python
# prints node table
layout.node_table()
```

```python
# prints edge table
layout.edge_table()
```

```python
# graph preview not yet implemented
layout.preview()
```

```python
# writes tables to nodes.csv and edges.csv
layout.write()
```
