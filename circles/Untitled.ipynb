```python
import networkx as nx
import numpy as np
from matplotlib import pyplot as plt
import ringity as rng
```

```python
circle = nx.cycle_graph(10)
```

```python
pos = nx.circular_layout(circle)
nx.draw(circle, pos = pos)
```

```python
dgm = rng.diagram(circle)
```

```python
rng.plot_dgm(dgm)
for coordinates in dgm:
    plt.plot([coordinates[0],coordinates[0]],[coordinates[0], coordinates[1]], 'gs--')
```

```python
rng.ring_score(dgm)
```

```python
perturbed_pos = pos
for i in pos:
    perturbed_pos[i][0] = pos[i][0] + np.random.rand()/2
    perturbed_pos[i][1] = pos[i][1] + np.random.rand()/2
```

```python
nx.draw(circle, pos=perturbed_pos)
```
