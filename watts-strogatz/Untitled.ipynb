```python
import networkx as nx
import numpy as np
from matplotlib import pyplot as plt
import ringity as rng
```

```python
G = nx.watts_strogatz_graph(20, 6, 0.2)
```

```python
pos = nx.circular_layout(G)
nx.draw(G, pos = pos, node_size=100)
plt.title('Watts-Strogatz-Graph: n=20, k=6, p=0.2')
plt.savefig('ws_circular.png')
```

```python
pos = nx.spring_layout(G)
nx.draw(G, pos = pos, node_size=100)
plt.text(-1, 0.8, r'ring score $\approx$ 0.56')
plt.savefig('ws_spring.png')
```

```python
dgm = rng.diagram(G)
```

```python
rng.plot_dgm(dgm)
for coordinates in dgm:
    plt.plot([coordinates[0],coordinates[0]],[coordinates[0], coordinates[1]], 's--', color=(0/255, 140/255, 160/255))

plt.text(10, 5, r'ring score $\approx$ 0.56')
plt.savefig('ws_persistence_diagram.png')
```

```python
rng.ring_score(dgm)
```

```python
G2 = nx.erdos_renyi_graph(20, 0.2)
```

```python
pos2 = nx.spring_layout(G2)
```

```python
nx.draw(G2, pos = pos2, node_size=100)
plt.text(-0.8, -0.7, r'ring score $\approx$ 0.05')
plt.title('Erdos-Renyi-Graph: n=20, p=0.2')
plt.savefig('er_spring.png')
```

```python
pos3 = nx.circular_layout(G2)
nx.draw(G2, pos = pos3, node_size=50)
plt.savefig('er_circular.png')
```

```python
dgm2 = rng.diagram(G2)
rng.plot_dgm(dgm2)
for coordinates in dgm2:
    plt.plot([coordinates[0],coordinates[0]],[coordinates[0], coordinates[1]], 's--', color=(0/255, 140/255, 160/255))
    
plt.text(25, 10, r'ring score $\approx$ 0.05')
plt.savefig('er_persistence_diagram.png')
```

```python
rng.ring_score(dgm2)
```
